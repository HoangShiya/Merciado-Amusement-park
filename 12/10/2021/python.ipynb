Bài 1: Viết chương trình tìm tất cả các số chia hết cho 7 nhưng không phải bội số của 5, nằm trong  đoạn 10 và 200 (tính cả 10 và 200). Các số thu được sẽ được in ra màn hình. 

```python
#Khởi tạo biến number trong khoảng 10-200
for x in range(10, 200):
  #Nếu số chia hết cho 7 và không phải bội của 5 thì xuất ra màn hình
  if (x % 7 == 0) and(x % 5 != 0):
    print(x)
```

Bài 2: Viết một chương trình tính giai thừa của một số nguyên dương n. Với n được nhập từ bàn  phím. Ví dụ, n = 8 thì kết quả đầu ra phải là 1*2*3*4*5*6*7*8 = 40320. 

```python
#Nhập giá trị đầu vào của n
NumberInput = int(input("Moi ban nhap gia tri n: "))

#Khởi tạo biến NumberOutput = 1 là biến xuất giá trị tính giai thừa của phép tính
NumberOutput = 1

#Có thể sử dụng vòng lặp để tính giai thừa hoặc sử dụng đệ quy
for i in range(NumberInput, 1, -1):
  NumberOutput *= i

#Xuất giá trị giai thừa của số nguyên dương n
print('Giá trị giai thừa của số n là ', NumberOutput)
```

Bài 4: Viết chương trình liệt kê tất cả các số nguyên tố nhỏ hơn n. Số nguyên dương n được nhập  từ bàn phím. 


```python
#Nhập giá trị đầu vào của n
NumberInput = int(input("Moi ban nhap gia tri n: "))

#Dùng vòng lặp để kiểm tra các số có phải số nguyên tố hay không hoặc dùng thuật toán sàng của Eratosthenes
for i in range(1, NumberInput):
  count = 0
  for x in range(1, i+1):
    if i % x == 0:
      count += 1
  if count == 2:
    print(i)

```

Bài 6: Viết chương trình liệt kê tất cả số nguyên tố có 5 chữ số.


```python

#Dùng loop để in các số nguyên tố có 5 chữ số, dùng step 2 để loại các số chẵn
for x in range(10001, 90000, 2):
  count = 0
  for i in range(1, x+1):
    if x % i == 0:
      count += 1
  if count == 2:
    print(x)
```

Bài 7: Viết chương trình tính tổng của các chữ số của môt số nguyên n. Số nguyên dương n được  nhập từ bàn phím. Ví dụ: n = 1234, tổng các chữ số: 1 + 2 + 3 + 4 = 10 

```python
#Nhập giá trị đầu vào của n
NumberInput = int(input("Moi ban nhap gia tri n: "))

#Khoi tao bien sum để tính tổng
sum = 0

#Thực hiện tách từng số rồi cộng lại với nhau
while (NumberInput > 0):
    sum = sum + NumberInput % 10;
    NumberInput = NumberInput // 10;

#Xuất giá trị tổng các chữ số ra màn hình
print(sum)
```
