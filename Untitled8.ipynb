```python
import pickle

"""
Class name: Person
Implementation: set name, Phonenumber, Email in outside
"""

class Person():
    def __init__(self, PERSON_NAME, PERSON_PHONE_NUMBER, PERSON_EMAIL):
        self.name = PERSON_NAME
        self.phone_number = PERSON_PHONE_NUMBER
        self.email = PERSON_EMAIL

    #Print 3 value Name, Phonenumber, Email in outside
    def printValueInOutside(self):
        print(f"Ho va ten: {self.name}\n")
        print(f"SDT: {self.phone_number}\n")
        print(f"Email: {self.email}\n")



"""
Class name: Student
Implementation: set student number and average mark in outside
Class Inheritance: Person class
"""

class Student(Person):
    def __init__(self, STUDENT_NUMBER, STUDENT_AVERAGE_MARK):
        self.number = STUDENT_NUMBER
        self.average_mark = STUDENT_AVERAGE_MARK

    #Print 2 value student number and average mark in outside
    def printValueInOutside(self):
        print(f"Number student: {self.number}\n")
        print(f"Diem trung binh: {self.average_mark}\n")



"""
Class name: Professor
Implementation: set salary in outside
Class Inheritance: Person class
"""

class Professor(Person):
    def __init__(self, PROFESSOR_SALARY):
        self.salary = PROFESSOR_SALARY

    #Print value salary in outside
    def printValueInOutside(self):
        print(f"Salary: {self.salary}")



#Input list incude 10 objects to 3 class
def inputData_for_list(ObjectList):
    for i in range(0, 10):
        #Initialization 3 variable for list
        ObjectList[i].NAME = input("Moi ban nhap ho va ten: ")
        ObjectList[i].PHONE_NUMBER = input("Moi ban nhap SDT: ")
        ObjectList[i].EMAIL = input("Moi ban nhap Email: ")

        #Person class: continue; Student class: Add student number, averager mark; Professor class: Add salary
        if ObjectList[i].checkClass() == "Person":
            continue
        elif ObjectList[i].checkClass() == "Student":
            ObjectList[i].STUDENT = input("Moi ban nhap ma HS: ")
            ObjectList[i].AVERAGE_MARK = input("Moi ban nhap diem trung binh: ")
        #Professor class
        else:
            ObjectList[i].SALARY = input("Moi ban nhap Salary: ")



"""
Name Function: Flash sort
@Complexity: O(n)
@Operating principle: divide and rule
@Based on algorithm: Bucket sort
_Implementation process:
    +STEP1: Find minimum value and largest element position in array: min_list, arr_list[max]
    +STEP2: Initialization L vector with m elements (corresponds to m classes. Approx, number of layers euqual to 0.45 x n)
    +STEP3: Count the number of elements of classes according to the rule: k = int((m - 1) * (arr_list[i] - min_list) / (arr_list[max] - min_list))
    +STEP4: Calculate the end position of the subclass j according to the formula: L[j] = L[j] + L[j - 1] (j increase from 1 to m-1)
"""
def flash_sort_increase(getList):
    bucket_number = round(0.45 * len(getList))

    #Initialization list to hold bucket
    bucket_list = []

    #Initialization 2 parameters for find min value and position with the max value
    bucket_max = 0
    bucket_min = getList[0]


    #Find max, min in list
    for i in range(1, len(getList)):
        if (getList[i] < bucket_min):
            getList[i], bucket_min = bucket_min, getList[i]
        if (getList[i] > bucket_max):
            getList[i], bucket_max = bucket_max, getList[i]

    #Count the elements in the bucket
    for i in range(0, len(getList)):
        k = round((bucket_number - 1) * (getList[i] - bucket_min) / (bucket_max - bucket_min))
        #Invalid elements will be ignored
        if bucket_list[k] is None:
            bucket_list[k] = 0
        bucket_list[k] += 1
        
    #Calculate the last index of each bucket
    for i in range(1, len(getList)):
        bucket_list[i] += bucket_list[i - 1]

    #Permutation (No need to use bucket memory)
    hold = getList[0]
    move = 0
    flash = 0
    para1 = 0
    para2 = 0
    para3 = 0

    while (move < len(getList[para1] - 1)):
        #If the element is already in the bucket, there is no need for permutation
        while (para3 > (bucket_list[para1] - 1)):
            para3 += 1

        flash = getList[para3]
        while (para3 != bucket_list[para1]):
            para1 = round((bucket_number - 1) * (getList[i] - bucket_min) / (bucket_max - bucket_min))
            hold = getList[para2 ]
            para2 = bucket_list[para1] - 1
            hold = getList[para2]
            flash = hold
            move += 1

    #Insertion sort
    for para3 in range(1, len(getList)):
        hold = getList[para3]
        i = para3 - 1
        while (i >= 0) and (getList[i] > hold):
            getList[i + 1] = getList[i]
            i -= 1
        getList[i + 1] = hold


def flash_sort_decrease(getList):
    bucket_number = round(0.45 * len(getList))

    # Initialization list to hold bucket
    bucket_list = []

    # Initialization 2 parameters for find min value and position with the max value
    bucket_min = 0
    bucket_max = getList[0]

    # Find max, min in list
    for i in range(1, len(getList)):
        if (getList[i] > bucket_min):
            getList[i], bucket_min = bucket_min, getList[i]
        if (getList[i] < bucket_max):
            getList[i], bucket_max = bucket_max, getList[i]

    # Count the elements in the bucket
    for i in range(0, len(getList)):
        k = round((bucket_number - 1) * (getList[i] - bucket_min) / (bucket_max - bucket_min))
        # Invalid elements will be ignored
        if bucket_list[k] is None:
            bucket_list[k] = 0
        bucket_list[k] += 1

    # Calculate the last index of each bucket
    for i in range(1, len(getList)):
        bucket_list[i] += bucket_list[i - 1]

    # Permutation (No need to use bucket memory)
    hold = getList[0]
    move = 0
    flash = 0
    para1 = 0
    para2 = 0
    para3 = 0

    while (move < len(getList[para1] - 1)):
        # If the element is already in the bucket, there is no need for permutation
        while (para3 > (bucket_list[para1] - 1)):
            para3 += 1

        flash = getList[para3]
        while (para3 != bucket_list[para1]):
            para1 = round((bucket_number - 1) * (getList[i] - bucket_min) / (bucket_max - bucket_min))
            hold = getList[para2]
            para2 = bucket_list[para1] - 1
            hold = getList[para2]
            flash = hold
            move += 1

    # Insertion sort
    for para3 in range(1, len(getList)):
        hold = getList[para3]
        i = para3 - 1
        while (i >= 0) and (getList[i] > hold):
            getList[i + 1] = getList[i]
            i -= 1
        getList[i + 1] = hold



def handle_File():

                        #----------------------------------------------------#
                        #             INITIALIZE CLASS PERSON                #
                        # ---------------------------------------------------#

    #Initialize list include object of class Person
    handle_File.List_Person = [Person() for i in range(0, 10)]
    inputData_for_list(handle_File.List_Person)

    #Sort list of people decrease with name
    flash_sort_decrease(handle_File.List_Person.name)

    #Save file with pickle
    handle_File.File_Person = open("Person", "wb")
    pickle.dump(handle_File.List_Person, handle_File.File_Person)
    handle_File.File_Person.close()

    #Read the file and output in screen
    handle_File.File_Person = open("Person", "rb")
    for i in range(0, 10):
        print(f"Information of person {i + 1} : {pickle.load(handle_File.File_Person[i]}")
    handle_File.File_Person.close()



                        #----------------------------------------------------#
                        #             INITIALIZE CLASS STUDENT               #
                        # ---------------------------------------------------#

    # Initialize list include object of class Student
    handle_File.List_Student = [Student() for i in range(0, 10)]
    inputData_for_list(handle_File.List_Student)

    # Sort list of people decrease with average
    flash_sort_decrease(handle_File.List_Student.average_mark)

    # Save file with pickle
    handle_File.File_Student = open("Person", "wb")
    pickle.dump(handle_File.List_Student, handle_File.File_Student)
    handle_File.File_Student.close()

    #Read the file and output in screen
    handle_File.File_Student = open("Student", "rb")
    for i in range(0, 10):
        print(f"Information of student {i + 1} : {pickle.load(handle_File.File_Student[i]}")
    handle_File.File_Student.close()


                        #----------------------------------------------------#
                        #            INITIALIZE CLASS PROFESSOR              #
                        # ---------------------------------------------------#

    # Initialize list include object of class Professor
    handle_File.List_Professor = [Student() for i in range(0, 10)]
    inputData_for_list(handle_File.List_Professor)

    # Sort list of people increase with salary
    flash_sort_increase(handle_File.List_Professor.salary)

    # Save file with pickle
    handle_File.File_Professor = open("Person", "wb")
    pickle.dump(handle_File.List_Professor, handle_File.File_Professor)
    handle_File.File_Professor.close()

    #Read the file and output in screen
    handle_File.File_Professorn = open("Professor", "rb")
    for i in range(0, 10):
        print(f"Information of professor {i + 1} : {pickle.load(handle_File.File_Professor[i]}")
    handle_File.File_Professor.close()





                            """"""""""""""""""""""""""""""""""""""""""""""""""""""""""""
                            """                                                     """"
                            """                   MAIN PROGRAM                      """"
                            """                                                     """"
                            """"""""""""""""""""""""""""""""""""""""""""""""""""""""""""
def main():
    handle_File()

if (__name__ = __main__):
    main()







```
