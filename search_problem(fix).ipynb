- Bước 4: Tìm kiếm x có xuất hiện trong danh sách trên hay không theo yêu cầu sau: o Nếu có thì trả về vị trí xuất hiện đầu tiên trong danh sách và thông báo số bước lặp  đã thực hiện để tìm được phần tử này.  
o Ngược lại, thông báo không tìm thấy.

```python
import random

def randomList(a, b, n):
  randList = random.sample(range(a, b), n)
  return randList


# Binary Search
def binarySearch(arr, x):
    left = 0
    right = len(arr) - 1
    while (left < right):
      mid = round((left + right) / 2)
      if x == arr[mid]:
          return mid

      elif x < arr[mid]:
          right = mid - 1

      elif x > arr[mid]:
          left = mid + 1

      else:
          return -1


# count Step
def countStep(arr, x):
    countStep = 0
    left = 0
    right = len(arr) - 1
    while (left < right):
      mid = round((left + right) / 2)
      if x == arr[mid]:
          break

      elif x < arr[mid]:
          right = mid - 1
          countStep += 1

      elif x > arr[mid]:
          left = mid + 1
          countStep += 1

    return countStep


###########################################
###                                     ###
###       Quick sort increase           ###
###                                     ###
###########################################
# STEP 1.1: partition_increase
def partition_increase(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot
 
    for j in range(low, high):
 
        # If current element is smaller than or
        # equal to pivot
        if arr[j] <= pivot:
 
            # increment index of smaller element
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
 
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
 
# STEP 1.2: quicksort_increase
def quicksort_increase(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
 
        # pi is partitioning index, arr[p] is now
        # at right place
        pi = partition_increase(arr, low, high)
 
        # Separately sort elements before
        # partition and after partition
        quicksort_increase(arr, low, pi-1)
        quicksort_increase(arr, pi+1, high)
# STEP 1.3: Sorted list
def sap_xep_danh_sach(x):
  n = len(x)
 
  quicksort_increase(x, 0, n-1)
  print(x)


###########################################
###                                     ###
###            MAIN PROGRAM             ###
###                                     ###
###########################################
def main(): 
  n = int(input("Moi ban nhap n so nguyen: "))
  a = int(input("Moi ban nhap so khoi dau: "))
  b = int(input("Moi ban nhap so ket thuc: "))
  x = int(input("Moi ban nhap gia tri x: "))

  # Initialize randList
  List = randomList(a, b, n)
  print("-----LIST BAN DAU-----")
  print(List)
  
  # List is sorted
  print("-----List sau khi được sắp xếp theo chiều tăng dần-----")
  sap_xep_danh_sach(List)

  # Function call 
  result = binarySearch(List, x)

  if result != -1:  
    print("Phan tu {} co vi tri dau tien la: {}".format(x, str(result)))
    print("So buoc lap da su dung la {}".format(countStep(List, x)))
  else:  
    print("Khong tim thay") 


if __name__ == "__main__":
  main()
```
