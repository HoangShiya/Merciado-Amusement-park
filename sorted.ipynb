```python
"""
::Name of class: Sorted
:param arr, low, high
:return: The sorted list (INCREASE)
STEP
//STEP 1: partition
@ This function takes last element as pivot, places
@ the pivot element at its correct position in sorted array, and places all smaller (smaller than pivot)
@ to left of pivot and all greater elements to right of pivot

//STEP 2: quicksort
@ The main function that implements QuickSort
@ arr[] --> Array to be sorted,
@ low  --> Starting index,
@ high  --> Ending index
@ Function to do Quick sort
"""
 
# STEP 1: partition
def partition(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot
 
    for j in range(low, high):
 
        # If current element is smaller than or
        # equal to pivot
        if arr[j] <= pivot:
 
            # increment index of smaller element
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
 
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
 
# STEP 2: quicksort
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
 
        # pi is partitioning index, arr[p] is now
        # at right place
        pi = partition(arr, low, high)
 
        # Separately sort elements before
        # partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)


##############################
##                          ##
##       Main program       ##
##                          ##
##############################
def __main__():
  # Initialize list include object of input
  list = [input("Moi ban nhap so thu {}: ".format(x+1)) for x in range(0, 10)]
  
  #list.sort(key = lambda x: x, reverse = False)

  # Initialize lenght of list
  n = len(list)

  quickSort(list, 0 , n-1)

  print("-----List sau khi được sắp xếp theo chiều tăng dần-----")
  print(list)

__main__()
```
